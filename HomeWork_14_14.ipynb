```python
import pandas as pd
import glob
import os

# –ü—É—Ç—å –∫ –ø–∞–ø–∫–µ
path = r'C:/Users/User/PycharmProjects/archive'

# –ù–∞–∑–≤–∞–Ω–∏—è —Ç–∞–±–ª–∏—Ü (–ø–æ –∏–º–µ–Ω–∞–º —Ñ–∞–π–ª–æ–≤)
table_names = {
    "Album.csv": "albums",
    "Artist.csv": "artists",
    "Customer.csv": "customers",
    "Employee.csv": "employees",
    "Genre.csv": "genres",
    "Track.csv": "tracks",
    "Playlist.csv": "playlists",
    "Invoice.csv": "invoices"
}

# –°–ø–∏—Å–æ–∫ —Ñ–∞–π–ª–æ–≤
all_files = glob.glob(path + "/*.csv")
# print("–ù–∞–π–¥–µ–Ω–Ω—ã–µ —Ñ–∞–π–ª—ã:", all_files)

# –°–æ–∑–¥–∞—ë–º —Å–ø–∏—Å–æ–∫ –¥–∞—Ç–∞—Ñ—Ä–µ–π–º–æ–≤
df_list = []

for filename in all_files:
    # –ü–æ–ª—É—á–∞–µ–º –∏–º—è —Ñ–∞–π–ª–∞
    file_name = os.path.basename(filename)
    table_name = table_names.get(file_name, "unknown")
    
    # print(f"–ß–∏—Ç–∞—é —Ñ–∞–π–ª: {file_name} –∫–∞–∫ —Ç–∞–±–ª–∏—Ü—É: {table_name}")

    # –ß–∏—Ç–∞–µ–º CSV
    df = pd.read_csv(filename, encoding='utf-8-sig', header=0)
    
    # –î–æ–±–∞–≤–ª—è–µ–º —Å—Ç–æ–ª–±–µ—Ü —Å –Ω–∞–∑–≤–∞–Ω–∏–µ–º —Ç–∞–±–ª–∏—Ü—ã
    df['table_name'] = table_name
    
    # –î–æ–±–∞–≤–ª—è–µ–º –¥–∞—Ç–∞—Ñ—Ä–µ–π–º –≤ —Å–ø–∏—Å–æ–∫
    df_list.append(df)

# –û–±—ä–µ–¥–∏–Ω—è–µ–º –ø–æ —Å—Ç—Ä–æ–∫–∞–º
combined_df = pd.concat(df_list, axis=0, ignore_index=True)

# –°–º–æ—Ç—Ä–∏–º —Ä–µ–∑—É–ª—å—Ç–∞—Ç
combined_df.head()
# print(f"üìä –†–∞–∑–º–µ—Ä –∏—Ç–æ–≥–æ–≤–æ–π —Ç–∞–±–ª–∏—Ü—ã: {combined_df.shape}")
```
