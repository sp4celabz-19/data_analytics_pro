# None / NaN

* **False**: Представляє логічне значення для "брехні". Може вказувати на відсутність чогось у логічних контекстах, але це все одно якесь значення.
* **Порожні колекції ([], {}, '')**
  * Порожні колекції, як-от порожні списки, словники та рядки, вважаються False у логічних контекстах, але це все одно якесь значення.
* **0**: Числовий нуль також вважається False у логічних контекстах, хоча це і справжнє число.
* **None**: Позначає відсутність значення. Python використовує ключове слово None для визначення нульових об’єктів і змінних.
* **NaN**: Спеціальне значення з плаваючою комою, яке використовується для представлення недійсних або невизначених числових результатів.

````python
x = None
print(x == None)   # True
print(x is None)   # True
print(bool(x)) # False (None is treated as False in boolean contexts)
````

````python
y_1 = float('nan')
y_2 = float('nan')

print(f'y_1 = {y_1}; y_2 = {y_2}')
print('y_1 == y_1 :', y_1 == y_1)      # False (NaN is not equal to anything, not even itself)
print('y_1 == y_2 :', y_1 == y_2)      # False (NaN is not equal to anything, not even itself)
print('bool(y_1)) :', bool(y_1))       # True (NaN is treated as True in boolean contexts)
````

````python
from math import nan, isnan

y_1 = float('nan')
y_2 = nan
print(f'y_1 = {y_1}; y_2 = {y_2}')
print('y_1 == y_1 :', y_1 == y_1)      # False (NaN is not equal to anything, not even itself)
print('y_1 == y_2 :', y_1 == y_2)      # False (NaN is not equal to anything, not even itself)
print('isnan(y_1) :', isnan(y_1))      # True
print('bool(y_1)) :', bool(y_1))       # True (NaN is treated as True in boolean contexts)
````

## Що таке функція?
* Функція — це блок коду, який виконує певне завдання.
* Функції допомагають упорядкувати код, покращити читабельність і зробити його модульним, дозволяючи повторне використання коду без його повторення.

## Навіщо нам функції?
* **Повторне використання коду**: ви можете викликати ту саму функцію в різних частинах коду, не переписуючи його.
* **Модульність**: функції дозволяють розбивати складні проблеми на менші, керовані завдання.
* **Покращена читабельність**: функції роблять код чистішим і легшим для розуміння.
* **Тестування**: Ви можете тестувати функції окремо, не тестуючи всю програму.
* Привчає до правильного мислення (Навіть якщо ви ще не пишете великі проєкти)
  * 🧠 Мислити функціями = мислити блоками логіки, як це робиться в реальних командах, проєктах і API.

## Назви функцій

## Positional Arguments vs. Keyword Arguments

[Невеличка стаття з прикладами](https://www.kdnuggets.com/2023/02/python-function-arguments-definitive-guide.html)


1. Позиційні аргументи:

  * Це аргументи, які потрібно передати функції в тому порядку, в якому визначено параметри.
  * Позиція кожного аргументу є критичною для правильної інтерпретації вхідних даних функцією.
  * Це найпоширеніший тип аргументів.
  * Якщо порядок змінено, функція може спрацювати несподівано.

```
def greet(first_name, last_name):
    print(f"Hello, {first_name} {last_name}!")

greet("John", "Doe")  # Correct order
# Output: Hello, John Doe!
```

2. Keyword Arguments:

  * Це аргументи, які передаються функції шляхом явного вказівки імені параметра разом із його значенням, незалежно від порядку.
  * Корисно для покращення чіткості та коли є багато параметрів.
  * Це дозволяє пропускати певні аргументи, використовуючи значення за замовчуванням або змінювати порядок аргументів.
  * Ви також можете комбінувати позиційні аргументи та аргументи за ключовими словами, але позиційні аргументи мають бути першими.



```
greet("John", last_name="Doe")  # Mixing positional and keyword arguments
# Output: Hello, John Doe!
```




Значення за замовчуванням в Keyword Arguments:

  * Функції можуть мати параметри зі значеннями за замовчуванням, які діють як аргументи ключових слів, якщо вони не вказані явно.

```
def greet(first_name="John", last_name="Doe"):
    print(f"Hello, {first_name} {last_name}!")

greet()  # Using default values
# Output: Hello, John Doe!
```



## Variable Number of Arguments

* Python дозволяє функціям приймати змінну кількість аргументів, тобто функція може обробляти різну кількість вхідних значень.
* Це корисно, коли ви заздалегідь не знаєте, скільки аргументів потрібно буде передати функції.

1. *args (Non-Keyword/Positional Variable Arguments):

  * *args дозволяє функції приймати змінну кількість позиційних аргументів. Усередині функції аргументи розглядаються як кортеж усіх переданих аргументів.

```
def sum_numbers(*args):
    return sum(args)

print(sum_numbers(1, 2, 3))  # Output: 6
print(sum_numbers(5, 10))    # Output: 15
```



````python
def print_lots(*args):
  print(args)

print_lots(1, 2, 3, 'Text')
````

2. **kwargs (Keyword Variable Arguments):
  * **kwargs дозволяє функції приймати змінну кількість аргументів ключового слова (пар ключ-значення).  Усередині функції kwargs розглядається як словник переданих аргументів ключового слова.

```
def print_user_info(**kwargs):
    for key, value in kwargs.items():
        print(f"{key}: {value}")

print_user_info(name="Alice", age=30)
# Output:
# name: Alice
# age: 30
```



````python
def print_user(**kwargs):
  print(kwargs)

print_user(name="Alice", age=30)
````

Ви можете використовувати як \*args, так і \*\*kwargs в одній функції, щоб прийняти будь-яку комбінацію позиційних і ключових аргументів.


```
def show_info(*args, **kwargs):
    print("Positional arguments:", args)
    print("Keyword arguments:", kwargs)

show_info(1, 2, 3, name="Alice", age=25)
# Output:
# Positional arguments: (1, 2, 3)
# Keyword arguments: {'name': 'Alice', 'age': 25}
```



````python
def show_info(*args, **kwargs):
  print("Positional arguments:", args)
  print("Keyword arguments:", kwargs)

show_info(1, 2, 3, name="Alice", age=25)
````

Випадки використання:

* *args: якщо потрібно, щоб функція обробляла різну кількість позиційних аргументів (наприклад, підсумовування невідомої кількості значень).
* **kwargs: коли вам потрібно гнучко обробляти іменовані аргументи, такі як опції конфігурації або параметри.

1. map():

* Призначення: Застосовує задану функцію до всіх елементів ітератора (наприклад, списку) і повертає об’єкт карти (який є ітератором).

````python
my_list = [1, 2, 3]
list(map(str, my_list))
````

````python
def square(a):
  return a**2

numbers = [1, 2, 3]
squares = list(map(square, numbers))
print(squares)  # Output: [1, 4, 9]
````

2. zip():

* Призначення: об’єднує елементи з двох або більше ітерованих (наприклад, списків або кортежів) у кортежі, поєднуючи елементи з однаковим індексом.
* Синтаксис: zip(iterable1, iterable2, ...)
* Випадок використання: коли вам потрібно поєднати елементи з кількох ітерацій.

````python
names = ['Alice', 'Bob', 'Charlse']
ages = [25, 30]
combined = list(zip(names, ages))
print(combined)  # Output: [('Alice', 25), ('Bob', 30)]
````

3. filter():

* Призначення: фільтрує елементи в ітерації, застосовуючи функцію, яка повертає True або False, і повертає лише ті елементи, для яких функція повертає True.
* Синтаксис: фільтр (функція, ітерація)
* Випадок використання: коли потрібно відфільтрувати елементи на основі умови.

````python
def is_even(a):
  if a%2 == 0:
    return True
  return False

numbers = [1, 2, 3, 4, 5]
even_numbers = list(filter(is_even, numbers))
print(even_numbers)  # Output: [2, 4]
````

## Lambda

## Namespaces

* Простір імен — це контейнер, де імена (ідентифікатори) зіставляються з об’єктами (наприклад, змінними та функціями).
* Це допомагає уникнути конфліктів імен, організовуючи змінні та функції в окремі простори.

### Types of Namespaces:

* Built-in Namespace: містить усі вбудовані імена, такі як функції (print(), len()) і винятки (наприклад, ValueError). Доступний globally.
* Global Namespace: містить імена, визначені на верхньому рівні сценарію або модуля. Ці імена доступні в усьому модулі.
* Local Namespace: містить імена, визначені всередині функції. Ці імена доступні лише в межах функції.

## Decorators

* Декоратори — це потужний інструмент у Python, який дозволяє змінювати або покращувати поведінку функцій або методів, не змінюючи їхній код.
* Вони часто використовуються для додавання таких функцій, як логування, контроль доступу або кешування.
* Декоратор — це функція, яка приймає іншу функцію як аргумент і повертає нову функцію, яка зазвичай розширює поведінку вихідної функції.
* Декоратори застосовуються до функції за допомогою синтаксису @decorator_name, розміщеного над визначенням функції.
* Ви можете застосувати кілька декораторів до однієї функції, склавши їх у стопку.

````python
def decorator_function(original_function):
    def wrapper_function():
        print("Wrapper executed before {}".format(original_function.__name__))
        return original_function()
    return wrapper_function

@decorator_function
def display():
    return "Display function executed."

display()
````

````python
def decorator_negative_numbers(original_function):
    def inner_function(x, y):
        new_x = max(0, x)
        new_y = max(0, y)
        return original_function(new_x, new_y)
    return inner_function

def surface_size_negative(x, y):
    return x * y

@decorator_negative_numbers
def surface_size(x, y):
    return x * y

print(surface_size_negative(x = 5, y = -5))
print(surface_size(x = 5, y = -5))
````

## Recursion

* Рекурсія — це техніка програмування, коли функція викликає саму себе, щоб вирішити проблему.
* Він зазвичай використовується для завдань, які можна розділити на менші схожі підзадачі.
* Рекурсивна функція зазвичай має два основні компоненти:
  * Base Case: умова, яка зупиняє рекурсію, щоб запобігти нескінченним циклам.
  * Recursive Case: частина функції, де вона викликає сама себе зі зміненими аргументами.
  

4. Переваги рекурсії:
  * Простота: рекурсивні рішення можуть бути більш простими та легшими для розуміння таких проблем, як обхід дерева, алгоритми пошуку тощо.
  * Чіткість коду: зменшує потребу у складних структурах циклу, роблячи код чистішим.

5. Недоліки рекурсії:
  * Продуктивність: рекурсивні функції можуть бути менш ефективними, ніж ітераційні рішення, через накладні витрати на численні виклики функцій.
  * Переповнення стека: якщо рекурсія занадто глибока, це може призвести до помилки переповнення стека (перевищення максимальної глибини рекурсії).


````python
5! = 5 * 4!
4! = 4 * 3!
````

````python
factorial(5) = 5 * factorial(4)
factorial(4) = 4 * factorial(3)
factorial(3) = 3 * factorial(2)
factorial(2) = 2 * factorial(1)
factorial(1) = 1
````

````python
def factorial(n):
    if n == 1:  # Base case
        return 1
    else:
        print(f'{n}*factorial({n} - 1)')
        k = n * factorial(n - 1)  # Recursive case
        print('Done for', n)
        return k

print(factorial(5))  # Output: 120
````

## Mutable parameters

````python
def append_to_list(value, my_list=[]):  # Mutable default parameter
    my_list.append(value)
    return my_list

print(append_to_list(1))
print(append_to_list(1))
````

````python
def append_to_list(value, my_list=None):
    if not my_list:
      my_list = []

    my_list.append(value)
    return my_list

append_to_list(value = 6)
````

* Значення за замовчуванням для my_list створюється один раз під час визначення функції, а не під час кожного виклику функції. Наступні дзвінки використовують той самий список.
* Щоб уникнути цієї проблеми, використовуйте None як параметр за замовчуванням і за потреби створіть новий змінний об’єкт у функції.

# Errors
Помилки в Python — це проблеми, які виникають під час виконання, що спричиняє зупинку програми. Їх можна згрупувати в дві основні категорії:




### Syntax Errors:
Виникає, коли парсер Python виявляє неправильний оператор.

Приклад:

```
if True print("Error")  # SyntaxError: missing colon
```

### Exceptions (Runtime Errors - Помилки виконання):

Виникають під час виконання програми.

Поширені типи винятків:
* NameError: Коли змінна не визначена.
* TypeError: Коли операція застосована до неправильного типу.
* ValueError: Коли функція отримує правильний тип, але невідповідне значення.
* IndexError: Під час спроби отримати доступ до індексу поза діапазоном списку або рядка.
* KeyError: При спробі отримати доступ до неіснуючого ключа в словнику.
* ZeroDivisionError: При спробі поділити на нуль.

Приклад:

```

```

## Обробка помилок
Use ***try, except*** blocks to handle exceptions gracefully.


```
try:
    x = 10 / 0
except ZeroDivisionError:
    print("Cannot divide by zero!")
```



````python
try:
  x = 10/0
  print(int('13/5'))
except:
  print('Error!')
````

````python
try:
  x = int('134[')
except ZeroDivisionError:
  print('Division by zero')
except ValueError:
  print('Text not right')
except:
  print('Error!')
````

## Виклик помилок

У Python ви можете створювати власні винятки, викликаючи помилки або визначаючи нові класи винятків. Це дозволяє більш ефективно обробляти конкретні ситуації у вашому коді.

Ви можете викликати вбудований виняток, використовуючи ключове слово ***raise***, щоб повідомити про помилку.



```
x = -5
if x < 0:
    raise ValueError("Negative numbers are not allowed!")
```



````python
def space(x, y):
  if x < 0 or y < 0:
    raise ValueError
  return x*y

space(2, -5)
````

## Створення власник помилок (погляд у майбутнє на бонусний урок)

Щоб створити власний виняток, визначте новий клас, який успадковує вбудований клас Exception (або будь-який інший конкретний клас винятків).



```
class CustomError(Exception):
    """Custom exception for specific use cases"""
    pass

# Raise the custom exception
raise CustomError("This is a custom error message!")
```




### Навіщо використовувати спеціальні помилки?

* Чіткість: спеціальні помилки роблять ваш код більш читабельним і допомагають виявити конкретні проблеми.
* Контроль: вони дозволяють вам обробляти певні умови, які можуть не відповідати існуючим типам помилок.

Ви можете обробляти спеціальні винятки так само, як і вбудовані, використовуючи ***try*** та ***except***.



```
try:
    raise CustomError("Something went wrong!")
except CustomError as e:
    print(f"Caught custom exception: {e}")
```

# 🕒 Модуль time

Модуль time надає функції для роботи з часом — вимірювання тривалості, затримки, поточний час тощо.

````python
import time
````

🔹 Основні функції

| Функція               | Опис                                             | Приклад                                                 |
| --------------------- | ------------------------------------------------ | ------------------------------------------------------- |
| `time.time()`         | Поточний час у секундах з 1970 року (Unix час)   | `time.time()` → `1717092404.45`                         |
| `time.sleep(seconds)` | Призупиняє виконання на вказану кількість секунд | `time.sleep(2)`                                         |
| `time.ctime(secs)`    | Перетворює секунди у читабельну дату             | `time.ctime(1717092404)` → `'Fri May 31 12:46:44 2025'` |
| `time.perf_counter()` | Високоточний таймер (для вимірювання тривалості) |                                                         |
| `time.process_time()` | Час процесора, витрачений програмою              |                                                         |


````python
start = time.time()
time.sleep(1.5)
end = time.time()
print("Час виконання:", end - start, "секунд")
````

````python
start = time.perf_counter()
time.sleep(1.5)
end = time.perf_counter()

print("Час виконання:", end - start, "секунд")
````

## ✅ **filter_definitions_by_part_of_speech**()

Description:
Write a function that filters definitions based on a given partOfSpeech. If no part is given, return all definitions.

```
def filter_definitions_by_part_of_speech(results, part='verb'):
    ...

```

Example:


```
filter_definitions_by_part_of_speech(word_data['results'], part='noun')

# Output:
SHENANIGAN (noun)
1. reckless or malicious behavior that causes discomfort or annoyance in others
2. the use of tricks to deceive someone (usually to extract money from them)
3. ...

```



````python
part = "noun"
word = word_data.get("word", "UNKNOWN").upper()

def filter_definitions_by_part_of_speech(results, part=None):
    filt_list = []

    for item in results:
        if part is None or item.get("partOfSpeech") == part:
            filt_list.append(item["definition"])

    return filt_list


definitions = filter_definitions_by_part_of_speech(word_data['results'], part=part)

if definitions:
    header = f"{word} ({part if part else 'all'})"
    print(header)
    for i, d in enumerate(definitions, start=1):
        print(f"{i}. {d}")
else:
    print(f"Part of speech '{part}' not found in any “partOfSpeech” of this dictionary")
````

## ✅ **extract_synonyms**() with lambda

Description:
Write a function that extracts all synonyms from the results, flattening them into one list. Use lambda + map() at least once. Remove duplicates.


---

Example:


```
extract_synonyms(word_data['results'])
# Output: ["devilment", "devilry", "deviltry", "mischief",...]

```


````python
def extract_synonyms_lambda(results):

    synonyms_all_lists = map(lambda x: x.get("synonyms", []), results)

    synonyms_one_lst = sum(synonyms_all_lists, [])

    return sorted(set(synonyms_one_lst))


unique_synonyms = extract_synonyms_lambda(word_data["results"])
print("Synonyms of the current word:", unique_synonyms)
````

## ✅ **most_common_POS**(word_lookup: dict) -> str

Description:
Analyzes the list of word usages in word_lookup["results"], collects all "partOfSpeech" entries, and returns the one that appears most frequently. Uses the Counter class from Python’s collections module.

Parameters:
* word_lookup: A dictionary containing data about a word (formatted like the WordsAPI response). Must include a "results" key with a list of dictionaries.

Returns:
* A string representing the most common part of speech (e.g. "verb", "noun")

Example:


```
most_common_POS(word_data)
# Output: 'noun'
```



````python
from collections import Counter


def most_common_part_of_speech(word_lookup):
    parts = [entry.get("partOfSpeech") for entry in word_lookup.get("results", []) if "partOfSpeech" in entry]

    if not parts:
        return None

    counter = Counter(parts)

    return counter.most_common(1)[0][0]

result = most_common_part_of_speech(word_data)

print(f'The most frequently "partOfSpeech" word is "{result}"')
````

## ✅ **summarize_word**(word_lookup: dict) -> None

Prints a human-readable summary of the word info.


Example:



````python
import time

def summarize_word(data):
    start = time.perf_counter()

    word = data.get("word", "Unknown")
    parts = {entry["partOfSpeech"] for entry in data.get("results", []) if "partOfSpeech" in entry}
    syllable_count = data.get("syllables", {}).get("count", "Unknown")
    pronunciation = data.get("pronunciation", {}).get("all", "Unknown")

    print("Output:\n")
    print(f"Word: {word}")
    print(f"Part(s) of Speech: {', '.join(sorted(parts))}")
    print(f"Syllables: {syllable_count}")
    print(f"Pronunciation: {pronunciation}")

    end = time.perf_counter()
    print(f"Function 'summarize_word' executed in {end - start:.4f} seconds")

summarize_word(word_data)
````
